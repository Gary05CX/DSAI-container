```python
from diffusers import StableDiffusionPipeline
import torch

from IPython.display import Image
```

```python
model_id: str = "Meina/MeinaUnreal_V3"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype = torch.float32, use_safeTensor = True, safety_checker = None)

pipe = pipe.to("cpu") # For Mac (testing)
```

```python
prompt = "neko"
negative_prompt = "Interloced fingers, three legs"
image = pipe(prompt, negative_prompt = negative_prompt).images[0]
image.save("/Code/output.png")
Image(filename = "/Code/output.png")
```
